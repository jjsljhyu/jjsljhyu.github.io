```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])
```

```python
model = keras.Sequential()
model.add(layers.Dense(64, activation="relu"))                     # add함수로 레이어를 추가함
model.add(layers.Dense(10, activation="softmax"))
```

```python
model.weights
```

빌드는 모델을 실행시킴
가중치 개수를 모름

```python
model.build(input_shape=(None, 3))
model.weights
```

입력의 노드 개수 3개 / 입력 형태는 모름

가중치를 시각화


summary -> 테이블 형태로 보여줌

파라미터 값 -> 빌드 64*3 +  64(바이어스 값)
              
                    64*10 + 10(바이어스 값)

```python
model.summary()
```

```python
model = keras.Sequential(name="example_model")
model.add(layers.Dense(64, activation="relu", name="first_layer"))
model.add(layers.Dense(10, activation="softmax", name="last_layer"))
model.build((None, 3))
model.summary()
```

input 레이어를 선언
input 쓸 때 shape 주의
(shape 쓸 때 앞,뒤 위치 바뀜)


```python
model = keras.Sequential()
model.add(keras.Input(shape=(3,)))
model.add(layers.Dense(64, activation="relu"))
model.summary()
```

```python
model.add(layers.Dense(10, activation="softmax"))
model.summary()
```
